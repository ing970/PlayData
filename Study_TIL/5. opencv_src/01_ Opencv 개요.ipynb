# Image란?

## Image는 width X height X channel 의 pixels로 이루어진 matrix

- width: 너비로 이미지의 가로 길이를 의미한다.
- height: 높이로 이미지의 세로 길이를 의미한다.

- pixel: image가 가지고 있는 값으로 width X height 개수만큼 존재한다. image를 표현하는 bit방식에 따라 값의 범위가 달라진다.

- 일반적인 Image는 unsigned integer 8bit로 표현
    - 0 ~ 255 로 표현
- bit수가 높아질 수록 표현할 수 있는 색의 범위가 증가한다.
- <span style='background-color:#fff5b1'>**shape은 넘파이 배열로 표기하기 떄문에 실제 데이터로 표기할때는 height X width로 표기한다**</span>
    - ex) (3,4)
- <span style='background-color:#fff5b1'>**채널은 각각의 3개의 채널로 구성되어있다.**</span>
    - red, green, blue로 구성되어있다.
    - 이 세가지를 RGB라 표현한다.
    - 3차원 배열로 표기된다.
    - ( h , w , c ) 로 표기되며, 보통 C(channel)는 3이다.
    - Channel의 개수가 1인 경우, 흑백색(gray scale) 이다.

## 색공간 변환
- **`cv2.cvtColor(src, code)`** 
    - <span style='background-color:#fff5b1'> **image 의 color space를 변환한다.**</span>
    - src: 변환시킬 이미지 (ndarray)
    - code
        - 변환시킬 색공간 타입 지정
        - cv2.COLOR_XXX**2**YYY 형태의 상수 지정(XXX를YYY로 변환)
            - cv2.COLOR_XXX2YYY 형태의 상수 지정(XXX를YYY로 변환)
            - cv2.COLOR_BGR2GRAY / cv2.COLOR_GRAY2BGR (BGR <-> GRAY)
            - cv2.COLOR_BGR2RGB  / cv2.COLOR_RGB2BGR  (BGR <-> RGB)
            - cv2.COLOR_BGR2HSV  / cv2.COLOR_HSV2BGR  (BGR <-> HSV)

> ### <span style='background-color:#fff5b1'>HSV</span>
> - Hue: 색상, 색의 종류
> - Saturation: 채도. 색의 선명도
> - Value : 명도. 밝기




```python
# img1(bgr mode) ==> rgb 모드로 변환
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) # (대상, flag)
plt.imshow(img2)
plt.show()
```

# 동영상 처리

## 동영상 읽기
- VideoCapture 클래스사용
    - 객체 생성
        - VideoCapture('동영상파일 경로'): 동영상파일  // <span style='background-color:#fff5b1'>**( 동영상파일 경로는 문자열 )**
        - VideoCapture(웹캠 ID): 웹캠  // <span style='background-color:#fff5b1'>**( 웹캠 ID는 정수열 )**
- VideoCapture의 주요 메소드
    - isOpened(): bool
        - 입력 대상과 연결되었는지 여부반환
    - <span style='background-color:#fff5b1'>read(): (bool, img)
        - Frame 이미지로 읽기
        - 반환값
            - bool: 읽었는지 여부
            - img: 읽은 이미지(ndarray)

#### 웹캠

```python
import cv2 

cap = cv2.VideoCapture(0)           # 정수: web 캠 연결. => 0부터 1씩 증가하는 값을 사용. 웹캠이 여러개인 경우 지정.
print(cap.isOpened())
while True:
    # 한개 Frame을 읽기
    success, image = cap.read()    # 한 프레임을 읽는다.
    if not success:
        print('프레임을 읽어오지 못함')
        break                       # 종료
    
    # 읽은 Frame을 출력
    image = cv2.flip(image, 1)      # 양수: 수평반전 // 0: 수직반전 // 음수: 수평/수직 모두 반전
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR => GRAY SCALE
    cv2.imshow('frame', image)
    if cv2.waitKey(1) == 27:        # 1000분에 1초마다 1프레임을 찍어낸다 // 27은 esc key ==> 종료
        break

        
cap.release()                       # 연결 닫기(해제)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## 동영상파일

-  <span style='background-color:#fff5b1'>**FPS (Frame Per Second)**</span>
    - 1초에 몇 Frame을 보여주는지. 동영상 파일에 설정되어있다.
    - 동영상파일의 FPS에 맞춰서(속도를 맞춰서) 플레이하려면 한 Frame 을 보여주고 FPS 맞는 시간간격만큼 기다려야 한다.
        - 기다리는 시간: 초/FPS

```python
cap_tmp = cv2.VideoCapture('images/wave.mp4')

FPS = cap_tmp.get(cv2.CAP_PROP_FPS)
DELAY1 = 1/FPS # 1 Frame 을 보여줄 시간
DELAY2 = 1000/FPS
print(FPS, f'{DELAY1}초, {DELAY2} 밀리초')


cap_tmp.release()
```

```python
import cv2
# VideoCapture 객체 생성 => 동영상파일 경로를 문자열로 전달.
cap = cv2.VideoCapture('images/wave.mp4') # 정수: web cam 연결, 문자열: 동영상파일 연결
print('연결성공여부:',cap.isOpened())

FPS = cap.get(cv2.CAP_PROP_FPS)
DELAY = int(1000/FPS) # waitKey()는 int값을 대기시간으로 받는다.

while True:
    
    success, image = cap.read() # 한 Frame 읽기
    if not success:
        print('읽기 실패')
        break # 종료
    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # 흑백으로 볼 시 image => gray scale
    
    cv2.imshow('frame', image)
    if cv2.waitKey(DELAY) == 27:
        break


cap.release()  # 연결 닫기(끊기)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### 동영상 저장
- capture(read)한 이미지 프레임을 연속적으로 저장하면 동영상 저장이 된다.
- VideoWriter 객체를 이용해 저장
    - **`VideoWriter(filename, codec, fps, size)`**
        - filename: 저장경로
        - codec
            - VideoWriter_fourcc 이용
            
        - fps
            - FPS (Frame Per Second) - 초당 몇 프레임인지 지정
        - size
            - 저장할 frame 크기로 원본 동영상이나 웹캠의 width, height 순서로 넣는다.
    - `VideoWriter().write(img)`
        - Frame 저장

```python
def func(*args, **kwargs):
    print(args, type(args), len(args))
    print(kwargs)
    
# func(1, 2, 3, 4, 5, 6, 7, 'a', 'b', 'c')
a = [1,2,3,4,5] # iterable
b = dict(a=1, b=2, c=3)



# func(*a)
# func(*'가나다라마바사')
# func(*'DIVX')

# func(*a, a=1, b=2, c=3)
func(*a, **b)
```

```python
# 웹캠에서 읽은 동영상을 파일로 저장
import cv2

cap = cv2.VideoCapture(0)

# 입력 영상의 size
_, img = cap.read()

# print(img.shape)
size = (img.shape[1], img.shape[0]) # width, height

# print(size)
codec = cv2.VideoWriter_fourcc(*'MJPG')
FPS = cap.get(cv2.CAP_PROP_FPS)
filepath = 'webcam_output.avi'

# VideoWriter 객체 생성
writer = cv2.VideoWriter(filepath, codec, FPS, size)

while True:
    success, image = cap.read()
    if not success:
        print('캡쳐실패')
        break
    
    cv2.imshow('frame', image)
    writer.write(image)
    if cv2.waitKey(1) == 27:
        break
        
cap.release()
writer.release()
cv2.destroyAllWindows()
cv2.waitKey(1)



```
